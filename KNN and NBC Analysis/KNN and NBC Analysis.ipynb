```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
```

```python
# Loading dataset
df = pd.read_csv("Bank.csv")
df.head()
```

```python
# Dropping ID and ZIP Code, columns
df2 = df.drop(columns=['ID', 'ZIP Code'])
df2.head()
```

```python
# Define features and target
X = df2.drop(columns=['Personal Loan'])
y = df2['Personal Loan']
```

```python
# Convert categorical variables with more than two categories into dummy variables
X = pd.get_dummies(X, drop_first=True)
```

```python
# Split data into training (60%) and validation (40%) sets
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
# Standardize the feature data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
```

```python
# Define the KNN classifier with k=1
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# Train the model
knn.fit(X_train_scaled, y_train)
```

```python
# Test different values of k to find the one that balances between overfitting and ignoring predictor information
best_k = 1
best_accuracy = 0

# Iterate over a range of k values (e.g., 1 to 20) to find the best k
for k in range(1, 21):
    # Define the KNN classifier with the current k
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Train the model
    knn.fit(X_train_scaled, y_train)
    
    # Predict on the validation data
    y_pred = knn.predict(X_valid_scaled)
    
    # Calculate the accuracy on the validation data
    accuracy = accuracy_score(y_valid, y_pred)
    
    # Update the best k if the current accuracy is better
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

best_k
```

```python
# Define the KNN classifier with the best k value (which was found to be 1)
knn = KNeighborsClassifier(n_neighbors=best_k)

# Training the model with the best k
knn.fit(X_train_scaled, y_train)

# Predicting on the validated data using the best k
y_valid_pred = knn.predict(X_valid_scaled)

# Generate the confusion matrix for the validation data
conf_matrix = confusion_matrix(y_valid, y_valid_pred)

print(conf_matrix)
```

```python
df_nbc = pd.read_csv("Bank.csv")
```

```python
# Dropping ID and ZIP Code, columns
df_nbc= df.drop(columns=['ID', 'ZIP Code'])
df_nbc.head()
```

```python
# Splitting the data into training (60%) and validation (40%) sets
X_nbc = df_nbc.drop(columns=['Personal Loan'])
y_nbc = df_nbc['Personal Loan']
X_train_nbc, X_valid_nbc, y_train_nbc, y_valid_nbc = train_test_split(X_nbc, y_nbc, test_size=0.4, random_state=42)
```

```python
# Using the pandas dataframe methods melt() and pivot()

# Melting the dataset with relevant columns only
melted_data = pd.melt(
    X_train_nbc.assign(Personal_Loan=y_train_nbc),
    id_vars=['CreditCard', 'Online', 'Personal_Loan']
)

# Create the pivot table from the melted data using 'Online' as columns, 'CreditCard' as rows, and 'Personal Loan' as secondary rows (count)
pivot_table_nbc = X_train_nbc.assign(Personal_Loan=y_train_nbc).pivot_table(
    index=['CreditCard', 'Personal_Loan'],
    columns='Online',
    aggfunc='size',
    fill_value=0
)

# Viewing the table
pivot_table_nbc
```

```python
# (b)
'''
Consider the task of classifying a customer who owns a bank credit card and is actively using
online banking services. Looking at the pivot table, what is the probability that this customer will
accept the loan offer? [This is the probability of loan acceptance (Loan = 1) conditional on
having a bank credit card (CC = 1) and being an active user of online banking services (Online =
1).]
'''
```

```python
# The customer owns a credit card (CC = 1) and is actively using online banking services (Online = 1)
credit_card = 1
online = 1

# Extract the count of customers with CreditCard = 1, Online = 1, and Loan = 1 (accepted)
loan_accepted = pivot_table_nbc.loc[(credit_card, 1), online]

# Extract the count of customers with CreditCard = 1, Online = 1 (total customers with CreditCard = 1 and Online = 1)
total_customers = pivot_table_nbc.loc[(credit_card,), online].sum()

# Calculate the conditional probability
probability_loan_acceptance = loan_accepted / total_customers

probability_loan_acceptance
```

```python
# c
'''
Create two separate pivot tables for the training data. One will have Loan (rows) as a function
of Online (columns) and the other will have Loan (rows) as a function of CC
'''
```

```python
# Create two separate pivot tables for the training data
# One with Loan (rows) as a function of Online (columns)
pivot_table_online = X_train_nbc.assign(Personal_Loan=y_train_nbc).pivot_table(
    index='Personal_Loan',
    columns='Online',
    aggfunc='size',
    fill_value=0
)

# One with Loan (rows) as a function of Credit Card (columns)
pivot_table_cc = X_train_nbc.assign(Personal_Loan=y_train_nbc).pivot_table(
    index='Personal_Loan',
    columns='CreditCard',
    aggfunc='size',
    fill_value=0
)

# Display both pivot tables to the user
print("Pivot Table 1 (Loan vs Online):\n")
print(pivot_table_online)
print("\nPivot Table 2 (Loan vs Credit Card):\n")
print(pivot_table_cc)
```

```python
# (d)
'''
d. Compute the following quantities [P(A ∣ B) means “the probability of A givenB”]:
i. P(CC = 1 ∣ Loan = 1) (the proportion of credit card holders among the loan acceptors)
ii. P(Online = 1 ∣ Loan = 1)
iii. P(Loan = 1) (the proportion of loan acceptors)
iv. P(CC = 1 ∣ Loan = 0)
v. P(Online = 1 ∣ Loan = 0)
vi. P(Loan = 0)
'''
```

```python
# Calculating the probabilities
P_CC_given_Loan_1 = pivot_table_cc.loc[1, 1] / pivot_table_cc.loc[1].sum()
P_Online_given_Loan_1 = pivot_table_online.loc[1, 1] / pivot_table_online.loc[1].sum()
P_Loan_1 = y_train_nbc.mean()  # Assuming y_train_nbc is a binary Series
P_CC_given_Loan_0 = pivot_table_cc.loc[0, 1] / pivot_table_cc.loc[0].sum()
P_Online_given_Loan_0 = pivot_table_online.loc[0, 1] / pivot_table_online.loc[0].sum()
P_Loan_0 = 1 - P_Loan_1

# Print the calculated probabilities
print("P(CC=1 | Loan=1):", P_CC_given_Loan_1)
print("P(Online=1 | Loan=1):", P_Online_given_Loan_1)
print("P(Loan=1):", P_Loan_1)
print("P(CC=1 | Loan=0):", P_CC_given_Loan_0)
print("P(Online=1 | Loan=0):", P_Online_given_Loan_0)
print("P(Loan=0):", P_Loan_0)
```

```python
# (e)
'''
 Use the quantities computed above to compute the naive Bayes probability
P(Loan = 1 ∣ CC = 1, Online = 1)
'''
```

```python
# Bayes rule: P(Loan = 1 | CC = 1, Online = 1) = P(CC = 1 | Loan = 1) * P(Online = 1 | Loan = 1) * P(Loan = 1) / P(CC = 1, Online = 1)

# Calculate P(CC = 1, Online = 1) for normalization (combining both loan and non-loan acceptors)
P_CC_1_Online_1 = (
    P_CC_given_Loan_1 * P_Online_given_Loan_1 * P_Loan_1 +
    P_CC_given_Loan_0 * P_Online_given_Loan_0 * P_Loan_0
)

# Calculate the naive Bayes probability P(Loan = 1 | CC = 1, Online = 1)
P_Loan_1_given_CC_1_Online_1 = (
    P_CC_given_Loan_1 * P_Online_given_Loan_1 * P_Loan_1
) / P_CC_1_Online_1

print("P(Loan=1 | CC=1, Online=1):", P_Loan_1_given_CC_1_Online_1)
```

```python
# (f)
'''
Compare this value with the one obtained from the pivot table in (b). Which is a more accurate
estimate?
'''
```

```python
# Comparing the probability obtained from the pivot table in (b) and the naive Bayes probability obtained in (e)

# Probability from pivot table (part b)
probability_pivot_table = probability_loan_acceptance

# Naive Bayes probability (part e)
probability_naive_bayes = P_Loan_1_given_CC_1_Online_1

# Calculate the difference between the two probabilities
difference = abs(probability_pivot_table - probability_naive_bayes)

print("Probability from Pivot Table:", probability_pivot_table)
print("Naive Bayes Probability:", probability_naive_bayes)
print("Difference between the two:", difference)
```

```python
'''
In this case, the pivot table probability (8.73%) is likely a more accurate estimate of loan acceptance for the given conditions within the dataset because it directly represents the observed data.
The naive Bayes estimate (9.52%) assumes conditional independence, which may not perfectly hold between features like CreditCard and Online usage
'''
```

```python
# Prepare the data for Naive Bayes model training
# Convert categorical variables with more than two categories into dummy variables
X_nbc_encoded = pd.get_dummies(X_nbc, drop_first=True)

# Split the encoded data into training and validation sets
X_train_nbc_encoded, X_valid_nbc_encoded, y_train_nbc_encoded, y_valid_nbc_encoded = train_test_split(X_nbc_encoded, y_nbc, test_size=0.4, random_state=42)

# Instantiate the Naive Bayes classifier
nb_classifier = GaussianNB()

# Train the model
nb_classifier.fit(X_train_nbc_encoded, y_train_nbc_encoded)

# Predict probabilities for the validation set
y_valid_probs = nb_classifier.predict_proba(X_valid_nbc_encoded)
```

```python
# Extract the probability P(Loan = 1 | CC = 1, Online = 1) from the model output for customers with CC = 1 and Online = 1
# Filter the validation set for customers with CC = 1 and Online = 1
filter_condition = (X_valid_nbc_encoded['CreditCard'] == 1) & (X_valid_nbc_encoded['Online'] == 1)
filtered_probs = y_valid_probs[filter_condition.values]

# Calculate the average probability of Loan = 1 for the filtered customers
average_prob_loan_1 = filtered_probs[:, 1].mean()

# Output the average probability
average_prob_loan_1
```
