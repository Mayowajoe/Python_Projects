```python
import numpy as np

# Defining the function f(x, y, z)
def f(x, y, z):
    return (np.exp(y) - np.log(x**8)) / (5 + np.sin(z))

# Lists of values for x, y, and z
x_values = [1, 2, 3, 4, 5]
y_values = [2, 3, 4, 5, 6]
z_values = [3, 4, 5, 6, 7]

# Computing the values for the function
results = []
for x, y, z in zip(x_values, y_values, z_values):
    result = f(x, y, z)
    results.append((x, y, z, result))

# Displaying the results
for x, y, z, value in results:
    print(f"f({x}, {y}, {z}) = {value}")
```

```python
# Define the function f(x)
def f(x):
    return np.exp(-x) / (4 * x)

# Define the range for x (from 0 to 50, inclusive)
x_values = np.arange(1, 51)  # Starting from 1 to avoid division by zero
y_values = f(x_values)

# Plot the graph of f(x)
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, marker='o')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = e^(-x) / (4 * x) for 0 ≤ x ≤ 50')
plt.grid(True)
plt.show()
```

```python
def areas(n):
    areas = []
    for i in range(1, n + 1):
        area = i**2
        areas.append(area)
    return areas

# Calculate areas for squares with lengths from 1 to 10
areas = areas(10)
print(areas)
```

```python
def calc_areas(n):
    areas = []
    for i in range(1, n + 1):
        area = i**2
        areas.append(area)
    return areas

# To evaluate the squares of the 10 variables
areas = calc_areas(10)
print(areas)
```
