```python
import pandas as pd
from tabulate import tabulate
df = pd.read_csv("Budget_-_2025_Budget_Ordinance_-_Positions_and_Salaries_20250125.csv")

# Query pay rate details for a given job title description.

def query_pay_rate_by_title(df, title_description):
    # Filter the data for the given title description
    results = df[df['TITLE DESCRIPTION'].str.contains(title_description, case=False, na=False)]
    
    # Select relevant columns
    selected_results = results[['DEPARTMENT DESCRIPTION', 'TITLE DESCRIPTION', 'BUDGETED PAY RATE', 'TOTAL BUDGETED AMOUNT']]
    
    return selected_results

def main():  
    title_to_query = input("Enter the job title to query: ")
    query_results = query_pay_rate_by_title(df, title_to_query)
    
    if not query_results.empty:
        # Use tabulate to print the results in table format
        print(tabulate(query_results, headers='keys', tablefmt='psql', showindex=False))
    else:
        print("No results found for the specified job title.")

if __name__ == "__main__":
    main()
```
