```python
# Question 5: Report the number of missing values for the variables MOFB, YOB, and AOR.
missing_mofb_value = d1['MOFB'].isna().sum()
missing_yob_value = d1['YOB'].isna().sum()
missing_aor_value = d1['AOR'].isna().sum()
print(missing_mofb_value, missing_yob_value, missing_aor_value)
```

```python
# Question 6: Create d2 data from d1 data by selecting variables such as RMOB, WI, RCA, Religion, Region, 
# AOR, HEL, DOB_CMC, DOFB_CMC, MTFBI, RW, NV, and RBMI.

d2 = d1[['RMOB', 'WI', 'RCA', 'Religion', 'Region', 'AOR', 'HEL', 'DOB_CMC', 'DOFB_CMC', 'MTFBI', 'RW', 'NV', 'RBMI']]
d2.head()
```

```python
# Question 7: Create a column of missing values corresponding to each row of the d2 data and name it as MISSING
d2['MISSING'] = d2.isna().sum(axis=1)
d2['MISSING']

# i re-ran the code after i was done removing the missing values
```

```python
# Question 8: Delete any missing values of the d2 data, and denote this by d3. This is a complete data
d3 = d2.dropna()
d3.isna().sum(axis=1)
```

```python
d3.head()
```

```python
del d3['MISSING'] #Removing MISSING column
```

```python
d3.head()
```

```python
# Question 9: Find the summary statistics of the d3 data frame
d3.describe()
```

```python
# Question 10: Sort the d3 data by variable names by alphabetical order. This will help you to easily locate the variable of your interest 
# as variables (columns) are alphabetically ordered.

d3 = d3[sorted(d3.columns)]
d3
```

```python
# Alternatively
d3.sort_index(axis=1, ascending=True)
```

```python
# Question 11: Add a new variable in the d3 data by finding the average of DOB_CMC, DOFB_CMC and MTFBI
d3['Avg_DOB_CMC_DOFB_CMC_MTFBI'] = d3[['DOB_CMC', 'DOFB_CMC', 'MTFBI']].mean(axis = 1)
```

```python
d3
```

```python
# Question 12: Create a variable “Newreligion” by recoding ‘1’ as ‘1’ and rest as ‘2’ from the Religion Variable.
d3['Newreligion'] = d3['Religion'].apply(lambda x: 1 if x == 1 else 2)
d3
```

```python
d3['Newreligion'].value_counts()
```

```python
# Question 13: Create BIC variable from MTFBI variable by recoding values of MTFBI less equal 9 months as 1, greater than 9 months 
# and less equal 24 months as 2, and greater than 24 is equal to 3

d3['BIC'] = d3['MTFBI'].apply(lambda x: 1 if x <= 9 else (2 if x <= 24 else 3))
d3
```

```python
d3['BIC'].value_counts()
```

```python
# Question 14: Now change categorical labels of BIC variable as ‘Early’ for 1, ‘Average’ for 2 and ‘Late’ for 3
d3['BIC'] = d3['BIC'].map({1: 'Early', 2: 'Average', 3: 'Late'})
d3
```

```python
# Question 15: Subset the d3 data by taking MTFBI variable greater than 10 with only rural population.
d3_subset =  d3[(d3['MTFBI'] > 10) & (d3['Region'] == 2)]
d3_subset
```

```python
# Question 16: Draw a boxplot for the MTFBI variable. Put appropriate labels for this graph. Make comments for this graph.
plt.figure(figsize=(8, 6))
plt.boxplot(d3['MTFBI'], vert=True, patch_artist=True)
plt.title('Boxplot of MTFBI Variable')
plt.xlabel('MTFBI')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
'''
The boxplot shows the distribution of the MTFBI variable. The median is near the lower end of the data, indicating that most values are small. 
The data is positively skewed, with a long tail extending towards higher values, and there are numerous outliers, including one extreme value around 1000. 
The interquartile range (IQR) is relatively small, suggesting that the middle 50% of data points are close together.
'''
```

```python
# Question 17:	Draw a histogram for the RCA variable. Put appropriate labels for this graph. Make comments for this graph.
plt.figure(figsize=(8, 6))
plt.hist(d3['RCA'], bins=15, color='red', edgecolor='black')
plt.title('Histogram of RCA Variable')
plt.xlabel('RCA')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
'''
The histogram for the RCA variable indicates a roughly uniform distribution, with a noticeable peak around 25. 
The data is spread across the range from 15 to 50, showing moderate variability.
This suggests that the RCA values are fairly dispersed, with no significant skewness.
'''
```

```python
# Question 18. Draw a bar chart for the Region variable. Put appropriate labels for this graph. Make comments for this graph.
region_counts = d3['Region'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(region_counts.index, region_counts.values, color='green', edgecolor='black')
plt.title('Bar Chart of Region Variable')
plt.xlabel('Region')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
'''
The Region variable chart shows that Regions 2 and 3 have the highest number of respondents with region 3 has the most populous, and Region 1 has the lowest. 
The distribution suggests an uneven population across regions.
'''
```

```python
# Question 19: Draw a pie chart for the Region variable. Put appropriate labels for this graph. Make comments for this graph
plt.figure(figsize=(8, 8))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Pie Chart of Region Variable')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.legend(title='Regions', labels=region_counts.index, loc='lower center', bbox_to_anchor=(0.5, -0.1), ncol=8)
plt.show()
```

```python
# Question 29.	For each split data in d4 write a single loop to find the mean, minimum, maximum, standard deviation of MTFBI.

# Note: We don't have any d4 dataset, guess it's an error. I will use d5 to solve this

for wi_value, subset in d5.items():
    mean_mtfbi = subset['MTFBI'].mean()
    min_mtfbi = subset['MTFBI'].min()
    max_mtfbi = subset['MTFBI'].max()
    std_mtfbi = subset['MTFBI'].std()
    print(f'WI = {wi_value}: Mean = {mean_mtfbi}, Min = {min_mtfbi}, Max = {max_mtfbi}, Std Dev = {std_mtfbi}')
```
